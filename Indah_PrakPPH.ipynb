## Data loading

**Reasoning**:
Import the pandas library and load the CSV file into a DataFrame, then display the first 5 rows.



```python
import pandas as pd

df = pd.read_csv("/Merged_Financial_Data.csv")
display(df.head())
```

## Data understanding

```python
df.info()
display(df.head())
display(df.describe())
```

## Data cleaning (optional)

### Subtask:
Periksa dan tangani nilai yang hilang atau tidak konsisten jika ada.


**Reasoning**:
Clean the 'pendapatan' column by removing non-numeric characters and converting it to a numeric type, then check for missing values across all columns and display their counts.



```python
df['pendapatan'] = df['pendapatan'].astype(str).str.replace(' MVC', '', regex=False)
df['pendapatan'] = pd.to_numeric(df['pendapatan'])

missing_values = df.isnull().sum()
display("Missing values per column:")
display(missing_values)
```

## Data analysis

### Subtask:
Lakukan analisis data sesuai dengan kebutuhan Anda. Ini bisa berupa perhitungan statistik, pengelompokan data, atau analisis tren.


**Reasoning**:
Calculate 'laba_sebelum_pajak', 'pajak', and 'laba_setelah_pajak' and then group the data by 'tahun' and 'skenario' to calculate the mean of 'laba_setelah_pajak'.



```python
df['laba_sebelum_pajak'] = df['pendapatan'] - df['beban_operasional'] - df['penyusutan']

df['pajak'] = 0
taxable_condition = (df['tahun'] < df['tax_holiday_awal']) | (df['tahun'] > df['tax_holiday_akhir'])
df.loc[taxable_condition, 'pajak'] = df.loc[taxable_condition, 'laba_sebelum_pajak'] * df.loc[taxable_condition, 'tax_rate']

df['laba_setelah_pajak'] = df['laba_sebelum_pajak'] - df['pajak']

df_analisis = df.groupby(['tahun', 'skenario'])['laba_setelah_pajak'].mean().reset_index()
display(df_analisis)
```

## Data visualization

### Subtask:
Buat grafik berwarna untuk memvisualisasikan hasil analisis. Pilih jenis grafik yang sesuai (misalnya, diagram batang, diagram garis, atau diagram lingkaran).


**Reasoning**:
Create a line plot to visualize the average net profit by year and scenario using the calculated `df_analisis`.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_analisis, x='tahun', y='laba_setelah_pajak', hue='skenario')
plt.title('Average Net Profit by Year and Scenario')
plt.xlabel('Year')
plt.ylabel('Average Net Profit')
plt.show()
```
